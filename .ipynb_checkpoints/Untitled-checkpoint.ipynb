```python
from flask import Flask,render_template,request,json,redirect, url_for, session, g
from flask_wtf.csrf import CSRFProtect
import os
import hashlib
from flask_sslify import SSLify
import mysql.connector
from mysql.connector.cursor import MySQLCursorPrepared

def run_query(query,params):
    conn = mysql.connector.connect(host='localhost',user='root',passwd='toor',db='project',use_pure=True)
    cursor = conn.cursor(buffered=True)
    cursor.execute(query,params)
    conn.commit()
    row = cursor.rowcount
    conn.close()
    return row
    
salt = "salttest"
app = Flask(__name__)
app.secret_key = os.urandom(24)
app.config.update(SESSION_COOKIE_SECURE=True,
                 SESSION_COOKIE_HTTPONLY=True,
                 SESSION_COOKIE_SAMESITE='Strict')
csrf = CSRFProtect(app)

def password_hash_generator(password):
    password = password+salt
    return hashlib.sha256(password.encode('utf-8')).hexdigest()

@app.route("/")
def main():
    return render_template('index.html')

@app.route('/showSignUp')
def showSignUp():
    return render_template('signup.html')                    

@app.route('/showsignin')
def showsignin():
    return render_template('signin.html')    

@app.route('/signUp', methods=['POST'])
def signUp():
    _name = request.form['inputName']
    _email = request.form['inputEmail']
    _password = request.form['inputPassword']
    
    if _name and _email and _password:                      
        search_user_query = "SELECT * FROM details WHERE Email = %s"
        params = (_email,)
        response = run_query(search_user_query,params)
        
        if response>0:
            return json.dumps({'html':'<span>User already exists!</span>'})
        else:
            _password = password_hash_generator(_password)
            create_user_query="INSERT INTO details (Fullame, Email, PasswordHash) VALUES (%s,%s,%s)"
            params = (_name,_email,_password)
            response = run_query(create_user_query,params)
            return redirect(url_for('showsignin'))
    
    else:
        return json.dumps({'html':'<span>Enter the required fields</span>'})
    
@app.route('/mainpage')
def mainpage():
    if g.user:
        return render_template('mainpage.html')

    return redirect(url_for('showsignin'))

@app.before_request
def before_request():
    g.user = None
    if 'user' in session:
        g.user = session['user']

@app.route('/getsession')
def getsession():
    if 'user' in session:
        return session['user']

    return 'User not logged in!'

@app.route('/dropsession')
def dropsession():
    session.pop('user', None)
    return redirect(url_for('showsignin'))

@app.route('/signIn', methods=['POST'])
def signIn():
    
    _email = request.form['inputEmail']
    _password = request.form['inputPassword']
        
    if _email and _password:
        search_query = "SELECT * FROM details WHERE Email = %s AND PasswordHash = %s"
        _password = password_hash_generator(_password)
    
        params = (_email,_password)
        cursor = run_query(search_query,params)
        if cursor:
            session.pop('user', None)
            session['user'] = _email
            return redirect(url_for('mainpage'))
        else:
            return json.dumps({'html':'<span>Credentials are wrong!!</span>'})
  
    else:
        return json.dumps({'html':'<span>Enter the required fields</span>'})
    
@app.route('/logout', methods=['POST'])
def logout():
    return redirect(url_for('dropsession'))
    
if __name__ == "__main__":
    app.run(ssl_context=('cert.pem','key.pem'))
    
    
    
```
